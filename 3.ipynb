```python
def n_tree(dx, dy):
    x = 0
    y = 0
    lx = len(m[0])
    ly = len(m)
    sum = 0
    while y + dy < ly:
        y += dy
        x += dx
        x %= lx
        sum += (m[y][x] == '#')
    return(sum)

n_tree(3, 1)
```
